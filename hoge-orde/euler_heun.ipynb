# Hoge-orde benadering 

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

## Euler

```python
e = math.e
n_lijst = [5,10,20,100]

def euler_method(num_steps,h):
    x = np.zeros(num_steps+1)
    y = np.zeros(num_steps+1)

    x[0] = 0
    y[0] = 1

    for step in range(num_steps):
        x[step+1] = h * (step + 1)
        y[step+1] = y[step] + h * x[step]
    return x,y

x, y = euler_method(100,0.01)
plt.plot(x,y)
plt.show()
```

## Heun

```python
def heun_method(num_steps,h):
    x = np.zeros(num_steps+2)
    y = np.zeros(num_steps+1)
    

    x[0] = 0
    y[0] = 2
    
    for step in range(num_steps+1):
        x[step+1] = h * (step+1)
        
    for step in range(num_steps):
        euler = y[step] + h * x[step]
        y[step+1] = y[step] + h * ((x[step] + euler) * 0.5)
    x = x[:-1]
    return x,y

x, y = heun_method(100,0.01)
plt.plot(x,y)
```
